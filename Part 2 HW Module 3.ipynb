```python
import os
import csv
```

```python
#Path to import from file location
py_poll = os.path.join("Resources", "pypoll-election_data.csv")
```

```python
#Set the variables

number_of_votes = 0

#all candidate votes = to 0
Stockham = 0
DeGette = 0
Doane = 0 

#all candidate percentages to 0
Stockham_percent = 0
DeGette_percent = 0
Doane_percent = 0 
```

```python
#Read in the csv file. When you write newline = ’’ it will get rid of the blank rows
with open(py_poll, newline= "") as csvfile:
    
    #Split the data in commas
    csvreader = csv.reader(csvfile, delimiter= ',')
    
    #so we can skip the header and just go to the first row
    header = next(csvreader)
    
    for row in csvreader:
        
        number_of_votes += 1
        
        #If the vote is for Stockham, then we add one vote to him
        if row[2] == "Charles Casper Stockham":
            Stockham += 1
        #If the vote is for DeGette, then we add one vote to her
        elif row[2] == "Diana DeGette":
            DeGette += 1
        #If the vote is for Doane, then we add one vote to him
        elif row[2] == "Raymon Anthony Doane":
            Doane += 1

#Store the amount of votes for each candidate
vote_result = {"Charles Casper Stockham":Stockham, "Diana DeGette":DeGette,"Raymon Anthony Doane":Doane}
     
#Calculate percentages of each candidate and round to 3 decimal places
Stockham_percent = round((Stockham / number_of_votes)*100, 3)
DeGette_percent = round((DeGette / number_of_votes)*100, 3)
Doane_percent =  round((Doane / number_of_votes)*100, 3)

Winner = max(Stockham, DeGette, Doane)
for result in vote_result:
    #print (result)
    #print (vote_result[result])
    
    if vote_result[result] == Winner:
       
        winner_name = result
    

#Winner = max(vote_result)
        
#Winner = max({"Charles Casper Stockham":Stockham, "Diana DeGette":DeGette,"Raymon Anthony Doane":Doane})

    #for vote_result in csvreader:
        
        #if float(Stockham_percent)> float(DeGette_percent):
            #Winner = "Charles Casper Stockham"
        #if (Stockham_percent)> float(Doane_percent):
            #Winner = "Charles Casper Stockham"
        #if float(DeGette_percent)> float(Stockham_percent):
            #Winner = "Diana DeGette"
        #if float(DeGette_percent)>float(Doane_percent):
            #Winner = "Diana DeGette"
        #if float(Doane_percent)> float(Stockham_percent):
            #Winner = "Raymon Anthony Doane"
        #if float(Doane_percent)> float(DeGette_percent):
            #Winner = "Raymon Anthony Doane"

            
            
print("Election Results \n" )
print("-------------------------- \n" ) 
print("Total number of votes: " + str(number_of_votes))
print("-------------------------- \n" )

print("Charles Casper Stockham: " + str(Stockham_percent) + "%" + "(" + str(Stockham) + ")")
print("Diana DeGette: " + str(DeGette_percent) + "%" + "(" + str(DeGette) + ")")
print("Raymon Anthony Doane: " + str(Doane_percent) + "%" + "(" + str(Doane) + ")") 

print("-------------------------- \n" )

print("Winner : " + str(winner_name))


#Just to check the number of votes for each candidate is correct!
#print(Stockham)
#print(DeGette)
#print(Doane)

#Check the percentages
#print(Stockham_percent)
#print(DeGette_percent)
#print(Doane_percent)

with open("PyPoll_output_Data", "w") as file:
    file.write("Election Results \n" )
    file.write("-------------------------- \n" ) 
    file.write("Total number of votes: " + str(number_of_votes))
    file.write("-------------------------- \n" )

    file.write("Charles Casper Stockham: " + str(Stockham_percent) + "%" + "(" + str(Stockham) + ")")
    file.write("Diana DeGette: " + str(DeGette_percent) + "%" + "(" + str(DeGette) + ")")
    file.write("Raymon Anthony Doane: " + str(Doane_percent) + "%" + "(" + str(Doane) + ")") 

    file.write("-------------------------- \n" )

    file.write("Winner : " + str(winner_name))
    
```
